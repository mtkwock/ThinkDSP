```python
# Part 3
spectrum.low_pass(cutoff=1000, factor=0.001)
spectrum.plot()
pylab.xlim([0, 600]) # So we can see if a fundamental frequency exists
thinkplot.show()
```

Plotting the spectrum gives the above graph.  There are peaks at around 110, 160, 210, 330, 440, 500, and 550.  The dominant frequency is 160Hz and the fundamental frequency is 110Hz.  However, because we are looking for harmonics, the most likely harmonic spacing looks to be around 55Hz instead.

```python
# Part 7
class SawtoothSignal(Sinusoid):
    def evaluate(self, ts):
        '''
        ts: Time of the signal
        Returns the evaluated height of a Sawtooth Signal:
           /|   /|
          / |  / |
         /  | /  |
        /   |/   |
        '''
        cycles = self.freq * ts + self.offset / PI2 # Copied from Triangle Wave, believe in the magic
        frac,_ = numpy.modf(cycles) # Determine what fraction through the cycle
        ys = normalize(unbias(frac), self.amp) # Scale to amplitude
        return ys

# Testing purposes to ensure that it works
sawtooth_signal = SawtoothSignal(freq=1, amp=1, offset=0)
sawtooth_wave = sawtooth_signal.make_wave(duration=10, start=0, framerate=50)
#sawtooth_wave.plot()

sawtooth_spectrum = sawtooth_wave.make_spectrum()
sawtooth_spectrum.plot()
pylab.show()
```

```python
# Part 10
# Looking for a signal with even and odd harmonics with a drop of like 1/(ff)
# Triangle and Square have odd only
# Sawtooth have both even and odd.  Looking this up, Sawtooth only has Odd
# Triangle Wave has 1/(ff): 8/PI2 sum(0, inf)

f = 500

# Details the 1/f^2 drop
triangle_signal = TriangleSignal(freq=f, amp=1, offset=0)

triangle_odd = triangle_signal.make_wave(duration=5, start=0, framerate=10000)
# Phase shifted by 1/4 of the period to make it an even function (sine to cosine)
triangle_even = triangle_signal.make_wave(duration=5, start=(0.25 / f), framerate=10000)

# triangle_odd.plot()
# triangle_even.plot()

# 
my_wave = triangle_odd + triangle_even
my_wave.write("Even and Odd f squared Dropoff.wav")
my_wave.plot()
pylab.xlim([0, 0.005])
pylab.title('Sum of Two triangle waves shifted a quarter of a phase')
pylab.show()
'The Fourier Transform is a linear operation, so adding means the final output will have both even and odd dropoffs of 1/ff'
```
